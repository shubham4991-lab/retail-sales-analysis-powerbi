```python
import pandas as pd

df1=pd.read_csv("online_retail_09_10.csv")
df2=pd.read_csv("online_retail_10_11.csv")

df1.head()
```

```python
df2.head()
```

```python
df = pd.concat([df1,df2], ignore_index=True)
df.shape
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['CustomerID'])
df.shape               
```

```python
df[df['Quantity']<0].shape[0], df[df['UnitPrice']<0].shape[0]
```

```python
df[df['Quantity']<0].head()
```

```python
df['InvoiceNo'].dtype,df['StockCode'].dtype,df['Description'].dtype,df['Quantity'].dtype,df['InvoiceDate'].dtype,df['UnitPrice'].dtype,df['CustomerID'].dtype,df['Country'].dtype
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CustomerID'] = df['CustomerID'].astype(int)

df['InvoiceNo'].dtype,df['StockCode'].dtype,df['Description'].dtype,df['Quantity'].dtype,df['InvoiceDate'].dtype,df['UnitPrice'].dtype,df['CustomerID'].dtype,df['Country'].dtype
```

```python
df['TransactionType']=df['Quantity'].apply(lambda x: 'Return' if x<0 else 'Sale')
df['TotalPrice']=df['Quantity'] * df['UnitPrice']

df.head()
```

```python
df.to_csv('clean_retail.csv',index=False)
```
